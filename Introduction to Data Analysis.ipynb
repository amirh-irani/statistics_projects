### Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate example data
mu = 10 # mean
sigma = 3 # standard
x = mu + sigma * np.random.randn(500)

num_bins = 50

fig, ax = plt.subplots()

# histogram
n, bins, patches = ax.hist(x, num_bins, density=True)

# ploting
ax.set_xlabel('X')
ax.set_ylabel('Probability density')
ax.set_title(f'Histogram of X: $\mu={mu}$, $\sigma={sigma}$')

fig.tight_layout()
plt.show()
```

### Practice with OOP with function objects

```python
def generate_data(mu: float, sigma: float, num_samples: int):
    '''
    mu: mean of the distribution
    sigma: standard deviaiton of distribution
    num_samples: number of samples to be generated
    '''
    return mu + sigma * np.random.randn(num_samples)
def histogram_plot(data: np.ndarray, num_bins: int):
    '''
    data: input data
    sigma: standard deviaiton
    num_bins: number of bins to plot histogram
    '''
    fig, ax = plt.subplots()
    # histogram
    n, bins, patches = ax.hist(data, num_bins, density=True)
    # ploting
    ax.set_xlabel('X')
    ax.set_ylabel('Probability density')
    ax.set_title(f'Histogram of X: $\mu={mu}$, $\sigma={sigma}$')
    fig.tight_layout()
    plt.show()
```

```python
data = generate_data(mu=10, sigma=3, num_samples=500)
histogram_plot(data=data, num_bins=50)
```

```python
data_new = generate_data(mu=100, sigma=10, num_samples=500)
histogram_plot(data=data_new, num_bins=50)
```

### Practice with OOP with class object

```python
class HistogramExample:
    def __init__(self, mu: float, sigma: float, num_samples: int,
        num_bins: int):
        """
        mu: mean of the distribution
        sigma: standard deviaiton of distribution
        num_samples: number of samples to be generated
        num_bins: number of bins to plot histogram
        """
        self.mu = mu
        self.sigma = sigma
        self.num_samples = num_samples
        self.num_bins = num_bins
    
    def _generate_data(self):
        return self.mu + self.sigma * np.random.randn(self.num_samples)
    
    def plot(self):
        data = self._generate_data()
        fig, ax = plt.subplots()
        
        # histogram
        n, bins, patches = ax.hist(data, self.num_bins,
        density=True)
        
        # ploting
        ax.set_xlabel('X')
        ax.set_ylabel('Probability density')
        ax.set_title(f'Histogram of X: $\mu={self.mu}$, $\sigma={self.sigma}$')
        fig.tight_layout()
        plt.show()
```

```python
hist = HistogramExample(mu=10, sigma=3, num_samples=500, num_bins=50)
hist.plot()
```
